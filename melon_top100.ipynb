```python
# 좋아요를 통틀어 가장 많이 받은 가수 : 방탄?
# 좋아요를 가장 많이 받은 곡 : 봄날? 
# 뉴진스의 곡은 몇 개나 올라와 있을까?
# 임영웅의 곡이 뉴진스의 곡보다 많을까?
```

```python
from melon_top100 import get_songs, get_like_count
```

```python
songs = get_songs()
```

```python
songs[:10]
```

```python
likes = get_like_count(
    [i["song_no"] for i in songs]
)
```

```python
likes
```

```python
import pandas as pd
```

```python
songs = pd.DataFrame(get_songs())
```

```python
songs.head()
```

```python
# songs의 song_no 칼럼에서 값을 하나씩 꺼내다가
# x라는 임시변수로 저장하고, likes[x]의 결과를
# 차곡차곡 songs["likes"]에 채워줘.

songs["likes"] = songs["song_no"].apply(lambda x: likes[x])
```

```python
songs.head()
```

```python
songs.sort_values("likes", ascending=False).head(10)
```

```python
# top100에 여러분이 좋아하는 가수(나는 뉴진스)의 곡이 몇 곡이나 올라와 있을까?
songs[songs["artist"] == "임영웅"]
```

```python
# 좋아요 수 기준으로 가장 많으면 1, 가장 적으면 100의 값을 가지는 rank라는 칼럼
songs["rank"] = songs["likes"].rank(ascending=False)
songs.head()
```

```python
songs[songs["rank"] > 95]
```

```python
# 한 앨범에서 가장 많은 곡이 Top100에 수록된 앨범
album_count = songs["album"].value_counts()
album_count
```

```python
# SELECT album_count.album, album_count.count, songs.artist 
# from album_count LEFT JOIN songs ON album_count.album = songs.album; 
```

```python
album_count = album_count.reset_index()
album_count
```

```python
album_count
```

```python
album_count = album_count.join(songs.set_index("album")["artist"], 
                               on="album", how="left").drop_duplicates()
```

```python
album_count = album_count[["album", "artist", "count"]]
album_count
```

```python
# %pip install koreanize_matplotlib
import koreanize_matplotlib
album_count.sort_values("count", ascending=True).plot.barh("album", figsize=(10,20))
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,8))
bars = ax.barh(album_count[:15][::-1]["album"], album_count[:15][::-1]["count"],
               height=0.6)
ax.bar_label(bars, padding=5)
plt.show()
fig.savefig("/mnt/c/users/hkit/desktop/asdf.png")
```

```python
# 모든 곡을 통틀어 좋아요를 가장 많이 받은 가수는?
songs.groupby("artist")["likes"].sum().sort_values(ascending=False)
```
