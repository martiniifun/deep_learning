```python
from keras import Sequential
from keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(1, 10, 500)
y = np.sin(X) * 10
```

```python
print(X)
print(y)
```

```python
plt.scatter(X, y)
plt.show()
```

```python
model = Sequential([
    Dense(units=150, activation="relu", input_dim=1),
    Dense(units=150, activation="relu"),
    Dense(units=150, activation="relu"),
    Dense(1)
])
```

```python
model.compile(optimizer="adam", loss="mse")
```

```python
# 학습
hist = model.fit(X, y, batch_size=1, epochs=100)
```

```python
plt.plot(X, y)
plt.scatter(X, model.predict(X), color="red")
```

```python
model.evaluate(X, y)
```
